# Question-1

```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

def compute_gradient_magnitude(gr_im, kx, ky):

    # using convolve2d to compute the gradients
    gradient_x = convolve2d(gr_im, kx, mode='same')
    gradient_y = convolve2d(gr_im, ky, mode='same')

    # using formula to calculate the gradients magnitude
    magnitude = np.sqrt(gradient_x**2 + gradient_y**2).astype(np.float64)
    return magnitude

def compute_gradient_direction(gr_im, kx, ky):
    
    # using convolve2d to compute the gradients
    gradient_x = convolve2d(gr_im, kx, mode='same')
    gradient_y = convolve2d(gr_im, ky, mode='same')

    # calculating the direction of the gradients using the arctan2 fucntin
    direction = np.arctan2(gradient_y, gradient_x).astype(np.float64)
    return direction

pic = cv2.imread(r'D:\uoe\2\cv\coursework\CA_2024\data/shapes.png')
gray = cv2.cvtColor(pic, cv2.COLOR_BGR2GRAY)


# horizontal and vertical kernels are as follows
kx = np.array([[-1, 0, 1],
               [-2, 0, 2],
               [-1, 0, 1]]).astype('uint8')

ky = np.array([[-1, -2, -1],
               [0, 0, 0],
               [1, 2, 1]]).astype('uint8')

# calling the functions
gradient_magnitude = compute_gradient_magnitude(gray, kx, ky)
gradient_direction = compute_gradient_direction(gray, kx, ky)

print("gradient_direction is \n{}\n".format(gradient_direction))
print("gradient_magnitude is \n{}".format(gradient_magnitude))

file_path = r'D:\uoe\2\cv\coursework\CA_2024\Answers/Answer_1_gradient_direction.npy'
np.save(file_path, gradient_direction)
file_path = r'D:\uoe\2\cv\coursework\CA_2024\Answers/Answer_1_gradient_magnitude.npy'
np.save(file_path, gradient_magnitude)
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

def generate_bovw_spatial_histogram(im, locations, clusters, divisions):
    # function for plotting the histograms and the images
    def plot_divisions_and_histograms(image, division, results, save_path=None):
        fig, axes = plt.subplots(2, 1, figsize=(10, 8))
        
        ax_image = axes[0]
        ax_hist = axes[1]

        ax_image.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        ax_image.set_title(f"Image with Divisions: {division}")

        for i in range(1, division[1]):
            ax_image.axvline(x=i * (image.shape[1] // division[1]), color='r', linestyle='--', linewidth=1)
        for j in range(1, division[0]):
            ax_image.axhline(y=j * (image.shape[0] // division[0]), color='r', linestyle='--', linewidth=1)

        colors = plt.cm.viridis(np.linspace(0, 1, num_clusters))
        hist = results[str(division)]
        ax_hist.bar(range(len(hist)), hist, color=colors)
        ax_hist.set_title(f"Histogram: {division}")
        ax_hist.set_xlabel('Bags of Words')
        ax_hist.set_ylabel('Frequency of visual words')
        # Save the plot before showing
        if save_path:
            plt.savefig(save_path)
        plt.show()
        plt.close()

    # initializing the results dictionary
    results = {}

    # generating bag of visual words spatial histogram for each division
    for division in divisions:
        h, w = im.shape[:2]
        div_y, div_x = division
        bin_size_y, bin_size_x = h // div_y, w // div_x

        histogram = np.zeros((div_y * div_x * num_clusters,), dtype=np.int64)

        for location, cluster_id in zip(locations, clusters):
            bin_x = int(location[0] // bin_size_x)
            bin_y = int(location[1] // bin_size_y)

            bin_idx = bin_y * div_x + bin_x
            cluster_idx = bin_idx * num_clusters + cluster_id

            histogram[cluster_idx] += 1

        results[str(division)] = histogram

        # this calls the above function for plotting
        # saving the histograms and the images plots along with the devision lines on the images
        save_folder = 'Answers'
        if not os.path.exists(save_folder):
            os.makedirs(save_folder)
        save_path = os.path.join(save_folder, f'Answer_2_visualisation_{division[0]}_{division[1]}.png')
        plot_divisions_and_histograms(im, division, results, save_path=save_path)

        # this is to display the 1d array
        print(f"Spatial histogram for division {division}:")
        print(histogram)
        print()

        # this is to save the 1d array
        np.save(os.path.join(save_folder, f'Answer_2_array_{division[0]}_{division[1]}.npy'), histogram)

    return results

# the following is the image path
image_path = r'D:\uoe\2\cv\coursework\Computer_Vision_Usama_730094899\data\books.jpg'
num_clusters = 12
divisions = [(1, 1), (2, 2), (2,3)]

# Load image
im = cv2.imread(image_path)
gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
sift = cv2.SIFT_create()
keypoints, descriptors = sift.detectAndCompute(gray, None)
locations = np.array([kp.pt for kp in keypoints], dtype=np.float32)

# Generate clusters
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
kmeans.fit(descriptors)
clusters = kmeans.predict(descriptors)

# Generate histogram
histogram_results = generate_bovw_spatial_histogram(im, locations, clusters, divisions)
```

```python
import numpy as np

def compute_rotation_matrix(points, theta):

    # converting the degrees to radians
    theta = (theta * np.pi) / 180.0
    
    # center of the points
    centroid = np.mean(points, axis=0)
    
    #translation matrices
    translate_to_origin = np.array([
        [1, 0, -centroid[0]],
        [0, 1, -centroid[1]],
        [0, 0, 1]
    ])
    
    translate_back = np.array([
        [1, 0, centroid[0]],
        [0, 1, centroid[1]],
        [0, 0, 1]
    ])
    
    # rotation matrix w.r.t to the origin
    rotation = np.array([
        [np.cos(theta), -np.sin(theta), 0],
        [np.sin(theta), np.cos(theta), 0],
        [0, 0, 1]
    ])
    
    #this will combine the matrix to rotate againts the centeroid
    rotation_matrix = translate_back @ rotation @ translate_to_origin
    
    return rotation_matrix.astype(np.float64)

points = np.load(r'D:\uoe\2\cv\coursework\CA_2024\data/points.npy')


theta = 90.0
rotation_matrix = compute_rotation_matrix(points, theta)


print("rotation_matrix is\n{}".format(rotation_matrix))
print("rotation_matrix_shape is {}".format(rotation_matrix.shape))
print("rotation_matrix_type is {}".format(rotation_matrix.dtype))


file_path = r'D:\uoe\2\cv\coursework\CA_2024\Answers/Answer_3_rotation_matrix.npy'
np.save(file_path, rotation_matrix)
```

```python
import torch
import torch.nn as nn
from torchvision import transforms, datasets
from ca_utils import ResNet, BasicBlock
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import transforms
from torchvision.datasets import ImageFolder
from torch.optim.lr_scheduler import StepLR
from torch.optim import AdamW
import os

def train_cnn(model, train_loader):
    num_epochs = 35
    lr = 0.01
    decay = 0.0005
    
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # this is to define the loss function, optimiser, and learning rate scheduler
    loss_function = torch.nn.CrossEntropyLoss()
    optimiser = AdamW(model.parameters(), lr=lr, weight_decay=decay)
    lr_scheduler = StepLR(optimiser, step_size=3, gamma=0.8)
    model.to(device)

    best_accuracy = 0.0
    best_model_path = None

    for epoch in range(num_epochs):
        model.train()
        total_loss = 0.0
        correct_predictions = 0
        total_samples = 0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            outputs = model(inputs)
            loss = loss_function(outputs, labels)
            optimiser.zero_grad()
            loss.backward()
            optimiser.step()

            total_loss += loss.item()

            predicted = torch.argmax(outputs, dim=1)
            total_samples += labels.size(0)
            correct_predictions += (predicted == labels).sum().item()

        # changing the learning rate
        lr_scheduler.step()

        # this will pring the training loss and accuracy at the end of each epoch
        epoch_loss = total_loss / len(train_loader)
        epoch_accuracy = correct_predictions / total_samples
        print("Epoch {}/{}, Loss: {:.4f}, Accuracy: {:.4f}".format(epoch + 1, num_epochs, epoch_loss, epoch_accuracy))


        # this will save the model weights if the accuracy is better than the previous one
        if epoch_accuracy > best_accuracy:
            best_accuracy = epoch_accuracy
            best_model_path = "data/weights_resnet.pth"
            torch.save(model.state_dict(), best_model_path)
            print("Best model weights have been saved at: {}".format(best_model_path))

    print("The training is complete!")


# this will create a resnet8 model and use 10 classes
num_classes = 10
model = ResNet(block=BasicBlock, layers=[1, 1, 1], num_classes=num_classes) 
n = model.fc.in_features
model.fc = nn.Linear(n, num_classes)


# the path for the images folders of test and train
data_directory_path = 'EXCV10/'

# dataset transformation
data_transforms = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5])
])

# loadging the data for images
train_loader = DataLoader(
    datasets.ImageFolder(os.path.join(data_directory_path, 'train'), transform=data_transforms),
    batch_size=32,
    shuffle=True
)

# this is the command to train the model
train_cnn(model, train_loader)
```

```python
import torch
import torch.nn as nn
from torchvision import transforms, datasets
from ca_utils import ResNet, BasicBlock
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import transforms
from torchvision.datasets import ImageFolder
from torch.optim.lr_scheduler import StepLR
from torch.optim import AdamW
import os
import numpy as np

def test_cnn(model, test_loader):
    # setting the model for evalutation mode
    model.eval()  
    predicted_labels = []
    correct_predictions = 0
    total_samples = 0

    # now we disable calculating the gradients for testing
    with torch.no_grad():  
        # we iterate over batches of testing data
        for images, labels in test_loader:
            outputs = model(images)
            i, predicted = torch.max(outputs, 1)
            predicted_labels.extend(predicted.numpy())
            total_samples += labels.size(0)
            correct_predictions += (predicted == labels).sum().item()
   
    accuracy = correct_predictions / total_samples * 100
    return np.array(predicted_labels), accuracy


# Data transformation for the testing set
test_transforms = transforms.Compose([
    # transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5])
])

# dictionary containing the test and val folders
data_directory_path = r'D:\uoe\2\cv\coursework\Computer_Vision_Usama_730094899\EXCV10/'

# loading the dataset for testing
test_dataset = datasets.ImageFolder(os.path.join(data_directory_path, 'val'), transform=test_transforms)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

# defining the model
num_classes = 10
model = ResNet(block=BasicBlock, layers=[1, 1, 1], num_classes=num_classes)
num_features = model.fc.in_features
model.fc = nn.Linear(num_features, num_classes)

# loading the pre-trained weights
model.load_state_dict(torch.load(r'D:\uoe\2\cv\coursework\Computer_Vision_Usama_730094899\data/weights_resnet.pth'))
model.eval()

# testing the model
predicted_labels, accuracy = test_cnn(model, test_loader)

# printing accuracy
print("Accuracy:", accuracy)
print(predicted_labels)

# saving accuracy
file_path = r'D:\uoe\2\cv\coursework\Computer_Vision_Usama_730094899/Answers/Answer_5_accuracy_of_model.npy'
np.save(file_path, accuracy)

file_path = r'D:\uoe\2\cv\coursework\Computer_Vision_Usama_730094899/Answers/Answer_5_predcited_labels.npy'
np.save(file_path, predicted_labels)
```

## Converting XML files into text format that is compatible for Yolo

```python
import os
import glob
import xml.etree.ElementTree as ET

def to_yolo(bbox, width, height):
    x = ((bbox[2] + bbox[0]) / 2) / width
    y = ((bbox[3] + bbox[1]) / 2) / height
    w = (bbox[2] - bbox[0]) / width
    h = (bbox[3] - bbox[1]) / height
    return [x, y, w, h]

def to_xml(bbox, width, height):
    w_half = (bbox[2] * width) / 2
    h_half = (bbox[3] * height) / 2
    xmin = int((bbox[0] * width) - w_half)
    ymin = int((bbox[1] * height) - h_half)
    xmax = int((bbox[0] * width) + w_half)
    ymax = int((bbox[1] * height) + h_half)
    return [xmin, ymin, xmax, ymax]

def convert_to_yolo_labels(input_path, output_path, image_path):
    classes = ['background_layer', 'with_mask', 'without_mask', 'mask_weared_incorrect']

    if not os.path.exists(output_path):
        os.mkdir(output_path)

    xml_files = glob.glob(os.path.join(input_path, "*.xml"))
    for xml_file in xml_files:
        filename = os.path.splitext(os.path.basename(xml_file))[0]
        image_file_path = os.path.join(image_path, f"{filename}.png")
        if not os.path.exists(image_file_path):
            print(f"{filename} image does not exist!")
            continue

        result = []

        tree = ET.parse(xml_file)
        root = tree.getroot()
        width = int(root.find("size").find("width").text)
        height = int(root.find("size").find("height").text)

        for obj in root.findall("object"):
            label = obj.find("name").text
            if label not in classes:
                classes.append(label)
            index = classes.index(label)
            bbox = [int(x.text) for x in obj.find("bndbox")]
            yolo_bbox = to_yolo(bbox, width, height)
            bbox_string = " ".join(map(str, yolo_bbox))
            result.append(f"{index} {bbox_string}")

        if result:
            with open(os.path.join(output_path, f"{filename}.txt"), "w", encoding="utf-8") as f:
                f.write("\n".join(result))

input_path_train = 'D:/uoe/2/cv/coursework/CA_2024/MaskedFace/train'
output_path_train = "D:/uoe/2/cv/coursework/CA_2024/MaskedFace/train/label"
image_path_train = "D:/uoe/2/cv/coursework/CA_2024/MaskedFace/train"

convert_to_yolo_labels(input_path_train, output_path_train, image_path_train)

input_path_val = 'D:/uoe/2/cv/coursework/CA_2024/MaskedFace/val'
output_path_val = "D:/uoe/2/cv/coursework/CA_2024/MaskedFace/val/label"
image_path_val = "D:/uoe/2/cv/coursework/CA_2024/MaskedFace/val"

convert_to_yolo_labels(input_path_val, output_path_val, image_path_val)
```

## Count Mask function and Mape calculator

```python
import os
import numpy as np
from collections import Counter

def count_masks(data_info):
    def calculate_mape(true_count_dict, pred_count_dict, category=['1', '2', '3']):
        error_values = []
        for i in category:
            error = abs(true_count_dict.get(i, 0) - pred_count_dict.get(i, 0))
            mape = 100 * error / max(1, true_count_dict.get(i, 1))
            error_values.append(mape)
        return np.mean(error_values)

    validation_labels_dir = data_info['validation_labels_dir']
    prediction_labels_dir = data_info['prediction_labels_dir']

    error_values = []
    predictions_data = []

    for label_file in os.listdir(validation_labels_dir):
        actual_count_data = Counter()
        pred_count_data = Counter()

        with open(os.path.join(validation_labels_dir, label_file)) as f:
            lines_read = f.readlines()
            data_labels = [line[0] for line in lines_read if line.strip()]
            actual_count_data.update(data_labels)

        if label_file in os.listdir(prediction_labels_dir):
            with open(os.path.join(prediction_labels_dir, label_file)) as f:
                lines_read = f.readlines()
                data_labels = [line[0] for line in lines_read if line.strip()]
                pred_count_data.update(data_labels)

        predictions_data.append(pred_count_data)
        error_values.append(calculate_mape(actual_count_data, pred_count_data))

    num_images_available = len(predictions_data)
    matrix_data = np.zeros((num_images_available, 3), dtype=np.int64)

    # Populate the matrix based on the predicted dictionaries
    for i, pred_count_data in enumerate(predictions_data):
        for j, key in enumerate(['1', '2', '3']):
            matrix_data[i, j] = pred_count_data.get(key, 0)
    
    return np.mean(error_values), matrix_data

# Example usage:
data_info = {'validation_labels_dir': "D:/uoe/2/cv/coursework/CA_2024/MaskedFace/val/label/", 'prediction_labels_dir': "D:/uoe/2/cv/coursework/CA_2024/yolov5/runs/detect/exp4/labels/"}
Mape ,  Matrix = count_masks(data_info)
print("Mean Absolute Percentage Error (MAPE) SCORE:", Mape)
print(Matrix.shape)
print("Data type of matrix_data:", Matrix.dtype)

# Saving the files
file_path = r'D:\uoe\2\cv\coursework\CA_2024/Answers/Answer_7_Mape.npy'
np.save(file_path, Mape)

file_path = r'D:\uoe\2\cv\coursework\CA_2024/Answers/Answer_7_Matrix.npy'
np.save(file_path, Matrix)


shape_path = r'D:\uoe\2\cv\coursework\CA_2024/Answers/Answer_7_Matrix_Shape.npy'
np.save(shape_path, Matrix.shape)

dtype_path = r'D:\uoe\2\cv\coursework\CA_2024/Answers/Answer_7_Matrix_Dtype.npy'
np.save(dtype_path, Matrix.dtype)

print("MAPE, Matrix, Matrix shape, and data type have been saved successfully.")
```
